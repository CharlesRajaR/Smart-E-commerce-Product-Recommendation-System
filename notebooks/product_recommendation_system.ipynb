```python
import os
import shutil

drive_path = '/content/drive/MyDrive/kaggle/datasets/amazon_reviws'
zip_source_path = '/content/amazon_review.zip'

# Create the directory in Google Drive if it doesn't exist
os.makedirs(drive_path, exist_ok=True)

# Copy the zip file to Google Drive
if not os.path.exists(os.path.join(drive_path, 'amazon_review.zip')):
    shutil.copy(zip_source_path, drive_path)
    print(f'Copied {zip_source_path} to {drive_path}')
else:
    print(f'amazon_review.zip already exists in {drive_path}')
```

Now that the `amazon_review.zip` file is saved to your Google Drive, you can access it across different Colab sessions. In future sessions, you would mount your drive and then copy the file from `drive_path` back to `/content/` if you need a fresh local copy, or directly extract from the Drive path if desired. For this session, we will proceed to extract the already present local file.

```python
import zipfile
import os

# Define the path to the zip file in Google Drive and the extraction directory
zip_file_path = os.path.join(drive_path, 'amazon_review.zip')
extract_dir = '/content/amazon_reviews_extracted'

# Create the extraction directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

# Extract the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)
    print(f'Extracted {zip_file_path} to {extract_dir}')

# List the contents of the extracted directory to verify
print('\nContents of extracted directory:')
for root, dirs, files in os.walk(extract_dir):
    for name in files:
        print(os.path.join(root, name))
    for name in dirs:
        print(os.path.join(root, name))
```
